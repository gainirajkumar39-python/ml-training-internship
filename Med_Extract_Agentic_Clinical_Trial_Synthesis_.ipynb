# **Data Loading and Initial Inspection**

```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/MUP_INP_RY24_P04_V10_DY20_Prv.csv")

# Basic inspection
print("Shape:", df.shape)
print("\nColumns:\n", df.columns)

df.head()
```

```python
# Example: selecting a target column (change based on your dataset)
target_column = df.columns[-1]   # last column as target (you can modify)

y = df[target_column]

print(y.head())
```

```python
# Features (X) and Target (y)
X = df.drop(columns=[target_column])
y = df[target_column]

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
print(X.columns[:10])  # first 10 columns
```

```python
# Check missing values
print("Missing values in X:", X.isnull().sum().sum())
print("Missing values in y:", y.isnull().sum())

# Handle missing values: Drop rows where the target variable is NaN
df_cleaned = df.dropna(subset=[target_column])

# Re-extract X and y from the cleaned DataFrame
X = df_cleaned.drop(columns=[target_column])
y = df_cleaned[target_column]

# Fill missing values in X (if any remain after dropping rows with NaN in y)
X = X.fillna(X.mean(numeric_only=True))
```

```python
print("\nTarget Distribution:\n")
print(y.value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)

print("Encoded labels sample:", y_encoded[:5])
print("Classes:", le.classes_)
```

```python
from sklearn.preprocessing import StandardScaler

# Select only numerical columns from X
X_numeric = X.select_dtypes(include=['number'])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_numeric)

print("Scaled shape:", X_scaled.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

r2_score = model.score(X_test, y_test)
print("R2 Score:", r2_score)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.4f}")
print(f"Root Mean Squared Error: {rmse:.4f}")
print(f"R-squared (R2) Score: {r2:.4f}")
```

```python
from sklearn.model_selection import cross_val_score

# Now using LinearRegression, and 'y' (continuous target) for cross-validation
scores = cross_val_score(model, X_scaled, y, cv=5, scoring='r2')

print("Cross-validation R2 Scores:", scores)
print("Mean R2 accuracy:", scores.mean())
```
